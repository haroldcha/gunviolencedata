## Cleaning

```python
# Cleaning #1: Adding 'year' column to help keep track of gun violence over year

df['datetime'] = pd.to_datetime(df["date"])
df['year'] = pd.DatetimeIndex(df['datetime']).year
```

```python
# Cleaning #2: Cleaning up geography name and replacing with 'city' and 'state' columns for city population

city = []
state = []
for i in df_2['Geographic Area'].values:
  area = i.split(" city, ")
  city.append(area[0])
  state.append(area[1])
  
# Cleaning #2.5: Removing commas from population values

df_2.replace(',','', regex=True, inplace=True)

df_2['city_or_county'] = city
df_2['state'] = state
city_pop_df = df_2.drop(columns='Geographic Area')
city_pop_df
```

```python
# Cleaning #3: Cleaning up geography name and replacing with 'state' columns for state population

state_pop_df = df_3.rename(columns={'Geographic Area': 'state'})
state_pop_df
```

```python
# Cleaning #4: Renaming 'city_or_county' values that have 'New York (Manhattan)' to 'New York'
# (for accurate comparsion of different cities' population vs gun violence incidents later)

df['city_or_county'] = df['city_or_county'].replace('New York (Manhattan)', 'New York')
```

```python
# Cleaning #5: Merging both population df's (state & city) with gun_violence df
# and then filtering down to 'city_pop' and 'state_pop column based on year/city and year/state, respectively

# @Team: This cell takes a bit of time to run; if you know a more efficient way
# to add population of city by year, please do so

# Merge with State population first
# This'll refer to state population by the years as (2013_x, 2014_x, etc.)
clean_df = df.merge(state_pop_df, how='left', on=['state'], sort=False)

# Merge with City population next
# This'll refer to city population by the years as (2013_y, 2014_y, etc.)
clean_df = clean_df.merge(city_pop_df, how='left', on=['city_or_county', 'state'], sort=False)

yr_idx = zip(clean_df['year'], clean_df.index)
pop_by_year_state = []
pop_by_year_city = []
for yr, idx in yr_idx:
    state = str(yr) + '_x'
    city = str(yr) + '_y'
    state_pop = clean_df.iloc[idx][state]
    city_pop = clean_df.iloc[idx][city]
    pop_by_year_state.append(state_pop)
    pop_by_year_city.append(city_pop)
clean_df['state_pop'] = pop_by_year_state
clean_df['city_pop'] = pop_by_year_city
clean_df = clean_df.drop(columns=['Census_x', 'Estimates Base_x', '2010_x', '2011_x', '2012_x',
                                  '2013_x','2014_x', '2015_x', '2016_x', '2017_x', '2018_x', '2019_x',
                                  'Rank', 'Census_y', 'Estimates Base_y', '2010_y', '2011_y', '2012_y',
                                  '2013_y','2014_y', '2015_y', '2016_y', '2017_y', '2018_y', '2019_y'])
```

```python
# Cleaning #6: Merging land area of states and calculating
# each state's population densities, then adding as 'pop_density' column

clean_df = clean_df.merge(df_4, how='left', on='state')
clean_df['pop_density'] = clean_df['state_pop']/clean_df['area (sq. mi)']
```

# Multiple Hypothesis Testing

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import abline_plot
import datetime
import math

import matplotlib.pyplot as plt
```

```python
# Determining whether or not the differences between the two evaluating groups are significant
# by comparing the groups' means and the individual value

def test_statistic(df, column):
    means_table = df.groupby(column).mean()
    if 1 not in means_table.index:
        diff = 0 - means_table.loc[0].values
    elif 0 not in means_table.index:
        diff = means_table.loc[1].values - 0
    else:
        diff = means_table.loc[1].values - means_table.loc[0].values
    
    return diff
```

```python
# Shuffling label of the group to test significance of the groups (A/B testing)

def simulated_test_stat(df, column):
    shuffled_labels = df.sample(frac=1, replace=True)[column]
    df["shuffled_labels"] = shuffled_labels.to_numpy()
    df = df.drop(column,axis=1)
    
    return test_statistic(df, 'shuffled_labels')   
```

```python
# Implementing A/B testing and bootstrapping to calculate p_value

def p_value(df, column, observed_difference, comparison):
    differences = []

    repetitions = 1000
    for i in np.arange(repetitions):
        new_difference = simulated_test_stat(df, column)
        differences = np.append(differences, new_difference) 
    if comparison=="greater":
        #indicates that higher values of the difference favor the alternative hypothesis that the treatment incidents were higher on average
        empirical_p = np.count_nonzero(differences >= observed_difference) / repetitions
    else:
        #indicates that lower values of the difference favor the alternative hypothesis that the treatment incidents were lower on average
        empirical_p = np.count_nonzero(differences <= observed_difference) / repetitions  
    return empirical_p
```

## 1. Are more populated cities associated with deadlier instances of gun violence?

Null hypothesis: In the U.S., the distribution of deadly instances of gun violence (according to our definition of 'deadly' above) is the same for cities that are populated and cities that are not. The difference in the sample is due to chance.

Alternative hypothesis: In the U.S., the instances of gun violence are deadlier for cities that are populated vs. cities that are not. 



```python
most_populous = df.loc[df['city_or_county'].isin(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'])]
most_populous['populous?'] = 1
least_populous = df.loc[df['city_or_county'].isin(['Charleston', 'Niagara Falls', 'Saginaw', 'Troy', 'Lakewood', 'Enid', 'Coral Gables', 'Cerritos', 'Texas City', 'Twin Falls'])]
least_populous['populous?'] = 0
populous = pd.concat([most_populous, least_populous])
populous_df = populous.groupby(['city_or_county', 'populous?']).sum()
populous_df['num_incidents'] = populous.groupby(['city_or_county', 'populous?']).count()['incident_id'].values
populous_df['num_victims'] = populous_df['n_killed'].values + populous_df['n_injured'].values
populous_df = populous_df.loc[:,['num_incidents', 'num_victims']]
populous_df = populous_df.reset_index('populous?')
populous_df
```

```python
victim_populous = populous_df.loc[:,['populous?', 'num_victims']]
test_victim_populous = test_statistic(victim_populous, 'populous?')
p_value(victim_populous, 'populous?', test_victim_populous, 'greater')
```

## Corrections

```python
# Controls FWER; increases decision-discoveries, but also increases chances of FP
def bonferroni(df, column, test, alpha):
    """
    Returns decisions on p-values using the Bonferroni correction.
    
    Inputs:
        p_values: array of p-values
        alpha_total: desired family-wise error rate (FWER = P(at least one false discovery))
    
    Returns:
        decisions: binary array of same length as p-values, where `decisions[i]` is 1
        if `p_values[i]` is deemed significant, and 0 otherwise
    """
    p_values = []

    for i in range(10):
        # test = test_statistic(df, column)
        p = p_value(df, column, test, 'greater')
        p_values.append(p)
    
    decisions = [(p_value <= alpha/len(p_values)) for p_value in p_values]
    return np.sum(decisions)/len(decisions)
```

```python
# Controls FDR; decrease FP, but also decreases decision-discoveries
def benjamini_hochberg(df, column, test, alpha):
    """
    Returns decisions on p-values using Benjamini-Hochberg.
    
    Inputs:
        p_values: array of p-values
        alpha: desired FDR (FDR = E[# false positives / # positives])
    
    Returns:
        decisions: binary array of same length as p-values, where `decisions[i]` is 1
        if `p_values[i]` is deemed significant, and 0 otherwise
    """
    p_values = []
    for i in range(10):
        # test = test_statistic(df, column)
        p = p_value(df, column, test, 'greater')
        p_values.append(p)
        
    p_list = np.array(p_values)
    p_list.sort()
    k = np.arange(len(p_values)) + 1
    pk = k*alpha/len(p_values)
    boolean_array = list(p_list <= pk)
    if sum(boolean_array) == 0:
        return 0
    p_max = np.max(p_list[boolean_array])
    decisions = [(p_value <= p_max) for p_value in p_values]
    return np.sum(decisions)/len(decisions)
```

### Using difference-in-difference for causal inference

Difference-in-difference is a statistical technique used to study the differential effects in natural experiments. The technique works by comparing a treatment group $T$ and a control group $C$ before and after an intervention. We first look at the slope before and after intervention of the control group. Then, we calculate the slope before and after intervention of the treatment. The difference between these two slopes, $Q$, is our "normal" difference between the two groups. That is our difference-in-difference estimate of the intervention effect is

$$
Q = (T_2 - T_1) - (C_2 - C_1)
$$

![difference in difference illustration](https://upload.wikimedia.org/wikipedia/commons/d/da/Illustration_of_Difference_in_Differences.png)

A major challenge with difference-in-difference is that in application, our calculated $Q$ can include confounding variables that occur before and after the intervention, so our implementation requires a **parallel trend assumption**. That is to say, in order to assume the treatment and control slopes are parallel in the absense of a treatment, we must account for all relevant confounders. This is difficult if we only calculate the naive difference of differences, so we will instead use a regression analysis. 

Regression models yield the same result when looking at individual-level data. Consider each unit of the natural experiment to have four values:

- $y$ is the value of interest
- $T$ is a binary value indicating whether they received the treatment
- $S$ is whether the unit was measured before or after the intervention
- $T \cdot S$ represents the interaction between time $\cdot$ treatment

See the following OLS model

$$
y = \beta_0 + \beta_1 S + \beta_2 T + \beta_3 T \cdot S \hspace{0.1cm} [+ \beta_{i+3} X_{i}] + \epsilon
$$

When running our regression analysis, the difference in difference estimate will be the calculated coefficient of the interaction term $T \cdot S$. This is the same as our naive calculation $Q$ because when calculating the conditional expectation of $y$, the population means are the same as the expected values from the regression.

Consider that each optimal coefficient corresponds to a difference in conditional expectations. For example,

$$
\hat{\beta_1} = \hat{E}(Y | S = 1, T = 0) - \hat{E}(Y | S = 0, T = 0)
$$

And in the first expectation term, by calculating the conditional expectation of $y$ given treatment before the intervention, we can obtain over all $y$ in that set the mean in the naive difference-in-difference estimator at the beginning of this section.

$$
\hat{E}(Y | S = 1, T = 0) = \frac{\hat{E}(y I(S = 1, T = 0))}{\hat{P}(S = 1, T = 0)}
$$

$$
= \frac{\sum_{i=1}^{n}{y_{i,S=1}I(i \in T = 0)}}{n_{T=0}} = \bar{y}_{T=0,S=1}
$$

We can apply these same steps to see that each $\beta_k$ corresponds to the means of units grouped by treatment and time. So by using a regression, we can easily include each confounder $X_i$ and still get an unbiased estimation of the intervention effects.

### Experiment design

Using difference-in-difference, we can see if our selected intervention, the 2014 Act 192, caused an increase in injuries + fatalities per incident. Our hypothesis is that if gun advocates can sue Philadelphia municipalities for gun legislation, more violent guns will be available and there will be a higher injury + death rate per gun incident.

Our value of interest $y$ is the number of deaths in a given incident. Our treatment value $T$ is a binary value indicating 1 if the incident occurred in South Philadelphia, PA or 0 if in Camden, NJ. We also have a value $I$ to indicate 1 if the violent incident occurred after the intervention, and 0 if it occurred before.

Since South Philadelphia and Camden are adjacent, we can exclude confounders such as weather, geography, etc. We'll be controlling for population density, median household income, unemployment rates, and the median age.

```python
def parse_age(age_string):
    if type(age_string) != str:
        return None
    
    ages = []
    age_split = []
    if '||' in age_string:
        age_split = age_string.split('||')
    elif '|' in age_string:
        age_split = age_string.split('|')
    else:
        age_split = [age_string]
    
    for age in age_split:
        if '::' in age:
            ages.append(int(age.split('::')[-1]))
        elif ':' in age:
            ages.append(int(age.split(':')[-1]))
        else:
            return None

    return np.mean(ages)
```

```python
phil_camden['mean_participant_age'] = phil_camden['participant_age'].apply(parse_age)
```

```python
did_df = phil_camden[['year', 'city_or_county', 'n_killed', 'n_injured', 'median_income', 'pop_density', 'unemployment', 'mean_participant_age']]
did_df = did_df.dropna()

did_df['T'] = 0
did_df.loc[did_df['city_or_county'] == 'Philadelphia', 'T'] = 1

did_df['I'] = 0
did_df.loc[did_df['year'] > 2014, 'I'] = 1

did_df['y'] = did_df['n_killed'] + did_df['n_injured']

did_df['T*I'] = did_df['T']

did_df = did_df[['y', 'T', 'I', 'T*I', 'median_income', 'pop_density', 'unemployment', 'mean_participant_age']]
did_df
```

Using OLS, we can find the equivalent value of the difference-in-difference estimate of the policy effect. That is, we can find the mean number of deaths before and after Act 192 in the Philadelphia treatment group, and the mean number of deaths before and after Act 192 in the Camden control group, and calculate the difference between these means. This is because our averages are just expected values from our estimator.

The reason we use OLS is to easily include our four confounders as well as calculating errors.

To do this, we can estimate the unknown parameters in $y = \beta_0 + \beta_1 T + \beta_2 I + \beta_3 TI + \beta_4 X_1 + \beta_5 X_2 + \beta_6 X_3 + \beta_7 X_4$ where $X_i$ is the $i^{th}$ confounder in our dataset.

```python
X = did_df.loc[:, did_df.columns != 'y']
Y = did_df['y']

model = sm.OLS(Y, X)
results = model.fit()
results.params
```

**Difference-in-Difference Using Fit OLS Model**

```python
def fit_OLS_model(df, target_variable, explanatory_variables, intercept = False):
    """
    Fits an OLS model from data.
    
    Inputs:
        df: pandas DataFrame
        target_variable: string, name of the target variable
        explanatory_variables: list of strings, names of the explanatory variables
        intercept: bool, if True add intercept term
    Outputs:
        fitted_model: model containing OLS regression results
    """
    
    target = df[target_variable]
    inputs = df[explanatory_variables]
    if intercept:
        inputs = sm.add_constant(inputs)
    
    fitted_model = sm.OLS(target, inputs).fit()
    return(fitted_model)
```
